# MSDS 400 Graphing Linear Systems

```python
import matplotlib.pyplot as plt
from numpy import linspace
```

The example shows how to plot a system of consistent equations:

3x + 10y = 115<br>
11x + 4y = 95


The function linspace divides the interval [-1,15] into 100 points. For each value in x, y1 and y2 will have a corresponding value based on the calculations shown below.

```python
x = linspace(-1, 15, 100)
y1 = 115.0/10.0 - (3.0/10.0)*x
y2 = 95.0/4.0 - (11.0/4.0)*x
```

Plots can be built up in layers.  What follows shows how to plot individual
colored lines.  plt.legend will associate the indicated equations to the 
preceding plt.plot statements.  The order of appearance must be the same. 
loc=3 places the legend in the lower left corner.

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'b')
plt.legend(['-5x+4y=-10', '3x+6y=-36'])
plt.title('Solving a System of Equations with a Unique Solution')
plt.show()
```

This example shows a plot of a system of inconsistent equations. 

2x - 3y = 6<br>
-4x + 6y = 8

It is not necessary to always specify colors for line plots particularly if no legend is involved. The computer will
select colors for you. Color codes may be found by entering `plot?` in the interpreter.  

The statement figure() separates the subsequent plot from the preceeding plot. Without it the two plots will be superimposed on each other. Take note of the order of statements. First, the variables being plotted are 
defined. The plot statements begin with `figure()` and end with `show()`.

```python
x = linspace(0, 6, 100)
y1 = 6.0/(-3.0)+(2.0/3.0) * x
y2 = 8.0/6.0 + (4.0/6.0) * x
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Showing an Inconsistent System of Equations')
plt.show()
```

This next example will show how to plot and solve a consistent system of non-linear equations using Python:

y = 2 - x<sup>2</sup><br>
y = x<sup>2</sup>

```python
x = linspace(0, 2, 100)
y1 = 2.0 - x**2
y2 = x**2
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Showing a Consistent System of Non-linear Equations')
plt.show()
```

**Note**: If a legend is used, the order of appearance of the plot statements must 
correspond to the terms used in the plt.legend statement.

```python
x = linspace(0, 100, 100)
y = x*x+x
y1 = y+1000.0
plt.figure()
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('Plot Showing Inconsistent Nonlinear Equations')
plt.plot(x, y, c='b')
plt.plot(x, y1, c='r')
plt.legend(['y=x**2+x', 'y=x**2+x+1000'])
plt.show()
```
