### Linear Equation Construction
Find the equation of the line L passing through the point (3, 7) and perpendicular to the line 5x - y = 4.

Note: The use of \n causes a line to be skipped before the text is printed.  The characters \n are otherwise ignored.

```python
x1= 5.0
y1= 4.0
x2= -10.0
y2= -2.0
slope= (y2-y1)/(x2-x1)
slope
```

The next example will show how to use the calculated slope to form the linear equation and calculate a result.  

```python
x= 1.0
y= y1 + slope*(x-x1)
print('\nValue of y if x is 1.0 equals {}'.format(y))
```

The next section shows how to assign a string to a variable for printing purposes.  The variable is called "equation" and it will be printed as a string.  This is a useful technique that will be used in later modules.

```python
equation= str('y = y1 + slope*(x-x1)')
print('Equation of a line is', equation)
print('x1 equals {} and y1 equas {}'.format(x1, y1))
```

The next example will show how to determine the break-even point graphically using Python.  Note that integers and floating point numbers can combined in the same equation.  The result is a floating point number.  The figure() statement separates the following plot from the previous plot. Two windows will appear with separate plots one behind the other.

A firm producing poultry feed finds that the total cost C(x) in dollars of producing sand selling x units is given by C(x) = 20x + 100. Management plans to charge $24 per unit for the feed. How many units must be sold for the firm to break even?

```python
x= linspace(0,50,100)
y= 20*x+100.0
z= 24*x
```

loc=2 places the legend in he upper left corner.  The order in which these statements appear is important.  legend() will associate 'cost' with the first statement and 'revenue' with the second. For more information about plot(), type the phrase  plot?  in the interpreter and enter.

```python
figure()  
plot (x,y)
plot (x,z)
legend (('cost','revenue'),loc=2)
title ('Breakeven Analysis')
show()
```
